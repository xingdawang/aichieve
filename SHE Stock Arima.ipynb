```python
data = history[['High']][-22*6:]

import statsmodels.api as sm
import itertools
import warnings

warnings.filterwarnings('ignore')

# setARIMA parameters p, d, q in range in [0,2). e.g[0, 1]
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 30) for x in list(itertools.product(p, d, q))]

# best params
best_pdp = None
best_seasonal_pdf = None
score = float('inf')

# go through all combination
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(data, 
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()
            print('ARIMA{} x {} - AIC:{}'.format(param, param_seasonal, results.aic)) # Akaike Information Criterion
            
            # get best params
            if results.aic < score:
                score = results.aic
                best_pdp = param
                best_seasonal_pdf = param_seasonal
        except:
            continue

print('Small AIC, Best pdf and PDQ:', score, best_pdp, best_seasonal_pdf)
```

```python
import statsmodels.api as sm

# bring data and prious determined parameters into the model
mod = sm.tsa.statespace.SARIMAX(data,
                                order=best_pdp,
                                seasonal_order=best_seasonal_pdf,
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])

# print diagnos result
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
pred_uc = results.get_forecast(steps=20*6)
pred_ci = pred_uc.conf_int()

pred_df = pred_uc.summary_frame()
pred_df
```

```python
# calculate Business Working Day
from pandas.tseries.offsets import BDay
from datetime import date

# start day from next working day
start_day = 1

for index, row in pred_df.iterrows():

    next_day = date.today() + BDay(start_day)
    start_day += 1
    pred_df.at[index, 'business_day'] = next_day.date()

pred_df.to_excel('she_prediction.xlsx')
pred_df
```

```python
import pandas as pd

# add previous day as start day
start_index = pred_df.index[0] - 1
value = data.iloc[-1]['High']
day = data.index[-1].date()
start_day = pd.DataFrame({
    'mean': value,
    'mean_se': 0,
    'mean_ci_lower': value,
    'mean_ci_upper': value,
    'business_day': day,
}, index=[start_index])

pred_df = pd.concat([start_day, pred_df])

# pred_ci

start_ci = pd.DataFrame({
    'lower High': value,
    'upper High': value
}, index=[start_index])

pred_ci = pd.concat([start_ci, pred_ci])
pred_ci
```
